### Modify census_1990_selected table

```python
# Make a copy to work with
census_1990_modified = census_1990_selected.copy()

# select only data from california
census_1990_modified = census_1990_modified[census_1990_modified['STATE'] == 'California']

# 1) Aggregate poverty level statistics (E07 columns) - calculate poverty rate
# E07001-E07012: Income above poverty level by age
# E07013-E07024: Income below poverty level by age
poverty_cols_above = [f'E07{str(i).zfill(3)}' for i in range(1, 13)]
poverty_cols_below = [f'E07{str(i).zfill(3)}' for i in range(13, 25)]

# total population above and below poverty
census_1990_modified['total_above_poverty'] = census_1990_modified[poverty_cols_above].sum(axis=1)
census_1990_modified['total_below_poverty'] = census_1990_modified[poverty_cols_below].sum(axis=1)
census_1990_modified['total_poverty_determined'] = census_1990_modified['total_above_poverty'] + census_1990_modified['total_below_poverty']

# poverty rate 
census_1990_modified['poverty_rate'] = (
    census_1990_modified['total_below_poverty'] / census_1990_modified['total_poverty_determined']
).fillna(0)

# Drop original E07 columns
e07_cols = [col for col in census_1990_modified.columns if col.startswith('E07')]
census_1990_modified = census_1990_modified.drop(columns=e07_cols)

census_1990_modified.head()
```

```python
# 2) Rename Race columns (E4S001-E4S005)
# E4S001: White
# E4S002: Black
# E4S003: American Indian, Eskimo, or Aleut
# E4S004: Asian or Pacific Islander
# E4S005: Other race

race_rename = {
    'E4S001': 'race_white',
    'E4S002': 'race_black',
    'E4S003': 'race_american_indian_eskimo_aleut',
    'E4S004': 'race_asian_pacific_islander',
    'E4S005': 'race_other'
}

census_1990_modified = census_1990_modified.rename(columns=race_rename)

census_1990_modified.head()
```

```python
# 3) Create occupancy status column (0 = vacant, 1 = occupied)
# EYP001: Occupied
# EYP002: Vacant

# Calculate occupancy rate: occupied / total housing units
census_1990_modified['total_housing_units'] = census_1990_modified['EYP001'] + census_1990_modified['EYP002']
census_1990_modified['occupancy_rate'] = (
    census_1990_modified['EYP001'] / census_1990_modified['total_housing_units']
).fillna(0)

# I'm going to keep these raw counts as well to preserve.
census_1990_modified = census_1990_modified.rename(columns={
    'EYP001': 'housing_occupied',
    'EYP002': 'housing_vacant'
})

census_1990_modified[['housing_occupied', 'housing_vacant', 'total_housing_units', 'occupancy_rate']].head(10)
```

```python
# 4) Rename year structure built columns (EX7001-EX7008)
# EX7001: 1989 to March 1990
# EX7002: 1985 to 1988
# EX7003: 1980 to 1984
# EX7004: 1970 to 1979
# EX7005: 1960 to 1969
# EX7006: 1950 to 1959
# EX7007: 1940 to 1949
# EX7008: 1939 or earlier

year_built_rename = {
    'EX7001': 'year_built_1989_to_1990',
    'EX7002': 'year_built_1985_to_1988',
    'EX7003': 'year_built_1980_to_1984',
    'EX7004': 'year_built_1970_to_1979',
    'EX7005': 'year_built_1960_to_1969',
    'EX7006': 'year_built_1950_to_1959',
    'EX7007': 'year_built_1940_to_1949',
    'EX7008': 'year_built_1939_or_earlier'
}

census_1990_modified = census_1990_modified.rename(columns=year_built_rename)

census_1990_modified[[col for col in census_1990_modified.columns if col.startswith('year_built_')]].head()
```

```python
# 5) Rename housing value columns (EZH001-EZH020)
# EZH001: Less than $15,000
# EZH002: $15,000 to $19,999
# EZH003: $20,000 to $24,999
# EZH004: $25,000 to $29,999
# EZH005: $30,000 to $34,999
# EZH006: $35,000 to $39,999
# EZH007: $40,000 to $44,999
# EZH008: $45,000 to $49,999
# EZH009: $50,000 to $59,999
# EZH010: $60,000 to $74,999
# EZH011: $75,000 to $99,999
# EZH012: $100,000 to $124,999
# EZH013: $125,000 to $149,999
# EZH014: $150,000 to $174,999
# EZH015: $175,000 to $199,999
# EZH016: $200,000 to $249,999
# EZH017: $250,000 to $299,999
# EZH018: $300,000 to $399,999
# EZH019: $400,000 to $499,999
# EZH020: $500,000 or more

# use a dictionary for this -- single call only in a funciton later
housing_value_rename = {
    'EZH001': 'home_value_less_than_15k',
    'EZH002': 'home_value_15k_to_19k',
    'EZH003': 'home_value_20k_to_24k',
    'EZH004': 'home_value_25k_to_29k',
    'EZH005': 'home_value_30k_to_34k',
    'EZH006': 'home_value_35k_to_39k',
    'EZH007': 'home_value_40k_to_44k',
    'EZH008': 'home_value_45k_to_49k',
    'EZH009': 'home_value_50k_to_59k',
    'EZH010': 'home_value_60k_to_74k',
    'EZH011': 'home_value_75k_to_99k',
    'EZH012': 'home_value_100k_to_124k',
    'EZH013': 'home_value_125k_to_149k',
    'EZH014': 'home_value_150k_to_174k',
    'EZH015': 'home_value_175k_to_199k',
    'EZH016': 'home_value_200k_to_249k',
    'EZH017': 'home_value_250k_to_299k',
    'EZH018': 'home_value_300k_to_399k',
    'EZH019': 'home_value_400k_to_499k',
    'EZH020': 'home_value_500k_or_more'
}

census_1990_modified = census_1990_modified.rename(columns=housing_value_rename)

census_1990_modified[[col for col in census_1990_modified.columns if col.startswith('home_value_')]].head()
```

```python
# final step: add a median home value column
# calculate median home value by finding which range contains the 50th percentile
def calculate_median_home_value(row):
    value_ranges = [
        ('home_value_less_than_15k', 0, 15000),
        ('home_value_15k_to_19k', 15000, 20000),
        ('home_value_20k_to_24k', 20000, 25000),
        ('home_value_25k_to_29k', 25000, 30000),
        ('home_value_30k_to_34k', 30000, 35000),
        ('home_value_35k_to_39k', 35000, 40000),
        ('home_value_40k_to_44k', 40000, 45000),
        ('home_value_45k_to_49k', 45000, 50000),
        ('home_value_50k_to_59k', 50000, 60000),
        ('home_value_60k_to_74k', 60000, 75000),
        ('home_value_75k_to_99k', 75000, 100000),
        ('home_value_100k_to_124k', 100000, 125000),
        ('home_value_125k_to_149k', 125000, 150000),
        ('home_value_150k_to_174k', 150000, 175000),
        ('home_value_175k_to_199k', 175000, 200000),
        ('home_value_200k_to_249k', 200000, 250000),
        ('home_value_250k_to_299k', 250000, 300000),
        ('home_value_300k_to_399k', 300000, 400000),
        ('home_value_400k_to_499k', 400000, 500000),
        ('home_value_500k_or_more', 500000, 600000)
    ]
    
    # Calculate total housing units
    total_count = sum(row[col] for col, _, _ in value_ranges)
    
    if total_count == 0:
        return 0
    
    # Find the median position (50th percentile)
    median_position = total_count / 2
    
    # Accumulate counts to find which range contains the median
    cumulative_count = 0
    for col, lower_bound, upper_bound in value_ranges:
        count = row[col]
        cumulative_count += count
        
        # If median position falls within this range
        if cumulative_count >= median_position:
            # Linear interpolation within the range
            previous_cumulative = cumulative_count - count
            position_in_range = median_position - previous_cumulative
            fraction_through_range = position_in_range / count if count > 0 else 0
            median_value = lower_bound + (fraction_through_range * (upper_bound - lower_bound))
            return median_value
    
    return 600000

census_1990_modified['median_home_value_1990'] = census_1990_modified.apply(calculate_median_home_value, axis=1)
home_value_cols = [col for col in census_1990_modified.columns if col.startswith('home_value_')]
census_1990_modified = census_1990_modified.drop(columns=home_value_cols)

census_1990_modified[['GISJOIN', 'STATE', 'median_home_value_1990']].head(10)

```

## Create & Save Final Dataset

```python
# merge the two datasets on GISJOIN
merged_census = pd.merge(census_1990_selected, census_201923_selected[['GISJOIN', '2019-23_value']], on='GISJOIN', how='inner')
merged_census
```

```python
# now final modification: add extra column that is ratio of 2019-2023 house value to 1990 average home value
merged_census['home_value_ratio_2019_23_to_1990'] = (
    merged_census['2019-23_value'] / merged_census['median_home_value_1990']
).replace([float('inf'), -float('inf')], 0).fillna(0)
merged_census.head()

```

```python
# export final merged dataset to CSV
merged_census.to_csv('data/merged_census_1990_2019_23_california.csv', index=False)
```

### Creating correlation matrices

```python
# select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['number'])
```

```python
# create correlation matrix in pyplot
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12, 10))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title('Correlation Matrix of Merged Census Data (1990 & 2019-23)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# necessary additional columns: 
# mean housing built year, percent white population

# 1) Calculate mean year housing was built
# Use midpoints of each year range weighted by the count of housing units
def calculate_mean_year_built(row):
    year_ranges = {
        'year_built_1989_to_1990': 1989.5,
        'year_built_1985_to_1988': 1986.5,
        'year_built_1980_to_1984': 1982,
        'year_built_1970_to_1979': 1974.5,
        'year_built_1960_to_1969': 1964.5,
        'year_built_1950_to_1959': 1954.5,
        'year_built_1940_to_1949': 1944.5,
        'year_built_1939_or_earlier': 1925  # Estimate for pre-1940 housing
    }
    
    total_weighted_years = 0
    total_units = 0
    
    for col, midpoint_year in year_ranges.items():
        count = row[col]
        total_weighted_years += midpoint_year * count
        total_units += count
    
    if total_units == 0:
        return 0
    return total_weighted_years / total_units

data['mean_year_built'] = data.apply(calculate_mean_year_built, axis=1)

# 2) Calculate percent white population
data['total_population'] = (
    data['race_white'] + 
    data['race_black'] + 
    data['race_american_indian_eskimo_aleut'] + 
    data['race_asian_pacific_islander'] + 
    data['race_other']
)

data['percent_white'] = (
    data['race_white'] / data['total_population'] * 100
).fillna(0)

data[['GISJOIN', 'mean_year_built', 'percent_white', 'total_population']].head(10)
```

```python
# update data object
numeric_data = data.select_dtypes(include=['number'])
```

```python
# create correlation matrix in pyplot -- REDO
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12, 10))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title('Correlation Matrix of Merged Census Data (1990 & 2019-23)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# identify pearson correlation coefficients and p-values for all numeric against 'home_value_ratio_2019_23_to_1990'

correlation_results = {}
for col in numeric_data.columns:
    if col != 'home_value_ratio_2019_23_to_1990':
        corr_coef, p_value = pearsonr(numeric_data[col], numeric_data['home_value_ratio_2019_23_to_1990'])
        correlation_results[col] = (corr_coef, p_value)
pd.DataFrame(correlation_results, columns=['Correlation_Coefficient', 'P_Value'])
```

```python
# rank by pvalue
sorted_correlation_results = sorted(correlation_results.items(), key=lambda x: x[1][1])
# show as dataframe
pd.DataFrame(sorted_correlation_results, columns=['Variable', 'Correlation_and_PValue'])
```

```python
# create a set of 9 scaterplots of 5 highest, 4 lowest p-values
top_5 = sorted_correlation_results[:5]
bottom_4 = sorted_correlation_results[-4:]
plt.figure(figsize=(15, 12))
for i, (var, (corr, pval)) in enumerate(top_5 + bottom_4):
    plt.subplot(3, 3, i+1)
    plt.scatter(numeric_data[var], numeric_data['home_value_ratio_2019_23_to_1990'], alpha=0.5)
    plt.title(f'{var}\nCorr: {corr:.2f}, P-value: {pval:.2e}')
    plt.xlabel(var)
    plt.ylabel('Home Value Ratio (2019-23 to 1990)')
plt.tight_layout()
plt.show()
```

```python
# final step: run corrs on ONLY percentage values (percent white, poverty rate, occupancy rate) and the average values (average home built year and average home value 1990) against the home value ratio
percentage_and_average_cols = ['percent_white', 'poverty_rate', 'occupancy_rate', 'mean_year_built', 'median_home_value_1990']
final_correlation_results = {}
for col in percentage_and_average_cols:
    corr_coef, p_value = pearsonr(numeric_data[col], numeric_data['home_value_ratio_2019_23_to_1990'])
    final_correlation_results[col] = (corr_coef, p_value)
    
# display in scatterplots
plt.figure(figsize=(12, 8))
for i, (var, (corr, pval)) in enumerate(final_correlation_results.items()):
    plt.subplot(2, 3, i+1)
    plt.scatter(numeric_data[var], numeric_data['home_value_ratio_2019_23_to_1990'], alpha=0.5)
    plt.title(f'{var}\nCorr: {corr:.2f}, P-value: {pval:.2e}')
    plt.xlabel(var)
    plt.ylabel('Home Value Ratio (2019-23 to 1990)')
plt.tight_layout()
plt.show()
```
